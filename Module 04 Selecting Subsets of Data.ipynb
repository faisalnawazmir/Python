# Selecting DataFrame rows

## Using the location indexers *.iloc* and *.loc*  for making the selection

```python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
college.head()
```

```python
pd.options.display.max_rows = 6
```

```python
# For selecting subsets of data, we use two "indexers" .iloc and .loc that respectively select the subset of data based 
# on location. The .iloc selects subset based on *integer location* and .loc selects susbset of data based on label of 
# the index.

#Both .iloc and .loc indexers can select data from a dataFrame or a Series
# Both indexers can be used to select row or columns or both.
# We can pass scalers, lists, or slices (e.g. 5:10) to the indexers
```

```python
# We can specify the integer location of the row with .iloc indexer
# We want 60th row

college.iloc[60]
```

```python
# Likewise, we can provide index label of the row we want and use .loc indexer
# We want the same row as above but now we provide the INSTNM variable's index label i.e. 'University of Alaska Anchorage'

college.loc['University of Alaska Anchorage']
```

```python
# We could also pass a list of rows to .iloc indexer 
# we want to subset (e.g 60th, 99th 3rd)
# Notice the use of double brackets to pass a list

college.iloc[[60, 99, 3]]
```

```python
# We can do the same things as above by passing the respective index labels and using .loc indexer
# Note we create a list of index labels first, and pass in to .loc indexer

labels = ['University of Alaska Anchorage',
          'International Academy of Hair Design',
          'University of Alabama in Huntsville']
college.loc[labels]
```

```python
# Alternatively we coudl pass a integer slice (e.g 99:102) to the .iloc indexer
# We want rows drom 99 to 102

college.iloc[99:102]
```

```python
# The label-based analogue to integer slice is to create a label slice
# we do this by specifying the start-label and the stop-label of the slice
# we use .loc indexer, and pass on to it the slice start:stop

start = 'International Academy of Hair Design'
stop = 'Mesa Community College'

college.loc[start:stop]
```

# Selecting DataFrame rows and columns simultaneously


```python
# We import the college data set
# Subset first 3 row and first 4 columns, using .iloc indexer

college = pd.read_csv('data/college.csv', index_col='INSTNM')
college.iloc[:3, :4]
```

```python
# We want to make the same selction as above but now with labels
# We specify all labels upto 'Amridge University' (i.e. row 3), 
# and all columns upto 'MENONLY' (i.e. columns 4 ) 
# We get the same result as before

college.loc[:'Amridge University', :'MENONLY']
```

```python
# Select all rows and two different columns (4rth and 6th) with .iloc indexer

college.iloc[:, [4,6]].head()
```

```python
# can make the same selection as above but based on column labels and by using .loc indexer

college.loc[:, ['WOMENONLY', 'SATVRMID']]
```

```python
# We can even select disjoint row and columns simultaneously by passing two lists to .iloc indexer
# One list is for subsetting rows i.e. [100, 200] and the other for subsetting columns [7, 15]

college.iloc[[100, 200], [7, 15]]
```

```python
# We can get the same result as above by passing lists of row- and columns-labels to .loc indexer

rows = ['GateWay Community College', 'American Baptist Seminary of the West']
columns = ['SATMTMID', 'UGDS_NHPI']

college.loc[rows, columns]
```

```python
# Next we want to subset the 6th row and 4rth columns from the right
# So lets look at the first 6 rows and all columns
# Note: The 4rth column from the right is PCTFLOANS (scroll to the left and see)

college.head(6)
```

```python
# To select 6th row, set row index =5 
# To select 4rth column from the right column index = -4
# pass both integer locations to .iloc indexer

college.iloc[5, -4]
```

```python
# Same result is achieved by passing the respective row and column labels to .loc indexer

college.loc['The University of Alabama', 'PCTFLOAN']
```

# Selecting data with both integers and labels

```python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
```

```python
# Let's look at columns of college dataFrame. Note 'CITY' is at index I=0 and UGDS_WHITE is at I=10
# and UGDS_UNKN is at I=18

college.columns
```

```python
#Note .columns.get_loc returns index I of the column.

col_start = college.columns.get_loc('UGDS_WHITE')
col_end = college.columns.get_loc('UGDS_UNKN') + 1

col_start, col_end
```

```python
# We want first 5 rows and every column between 10th and 19 columns, both exclusive

college.iloc[:5, col_start:col_end]
```

```python
row_start = college.index[10]
row_end = college.index[15]
row_start, row_end
```

```python
college.loc[row_start:row_end, 'UGDS_WHITE':'UGDS_UNKN']
```

# Scalar selection

```python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
cn = 'Texas A & M University-College Station'
college.loc[cn, 'UGDS_WHITE']
```

```python
# A better way is to use .at indexer when we have a scalar selection. It is faster

college.at[cn, 'UGDS_WHITE']
```
