# Working with Operators on a DataFrame
## *Applying the same transformation to all variables*

```python
college = pd.read_csv('data/college.csv')
college.head()
```

```python
college + 5  # will give error because not all columns are numeric
```

```python
# We need to extract numeric columns
##   Suppose we extract all colums that give fraction of students for different races
##   These columns start with prefic 'UGDS_'
##   So we use filter method with parameter like= 'UGDS_'

college = pd.read_csv('data/college.csv', index_col = 'INSTNM')
college_ugds_ = college.filter(like='UGDS_')
college_ugds_.head()
```

```python
# Now we can add a small faction such as .00501 to all columns
# We do it with a single dataframe operation

(college_ugds_ + .00501).head()

# Note: the addition of .00501 is performed on all columns of the DataFrame in one go.
```

```python
# Likewise, we can perform floor division to convert to nearest whole number percentages
(college_ugds_ + .0051)//.01
```

```python
# We divide by 100 to complete the rounding exercise
college_ugds_op_round = (college_ugds_ + .00501) // .01 / 100
college_ugds_op_round.head()
```

```python
# DataFrames also have methods equivalent of operators. See below
# Instead pf operators we perform the same operations as above with methods

college_ugds_op_round_methods = college_ugds_.add(.00501) \
                                            .floordiv(.01) \
                                            .div(100)

# Check if the dataFrame coreated with operators is the same as the one created with methods

college_ugds_op_round_methods.equals(college_ugds_op_round)
```

# Comparing Missing Values

```python
# This recipe shows how how to handle missing values.
#    We will first learn to make elementwise comparisons of dataFrame
#    In particular, we will learn to compare elements to np.nan 
#    
college =pd.read_csv('data/college.csv', index_col= 'INSTNM')
college_ugds_ = college.filter(like='UGDS_')

# college_ugds_.head()
```

```python
# Testing whether all values of all columns equal a scalar constant (say, .0019)

college_ugds_ == .0019

# Obviously this is comparison is false
```

```python
# Elementwise comparison ofa dataFrame to itself

college_self_compare = college_ugds_ == college_ugds_
college_self_compare.head()

# You would expect this comparison to return 'True' of each element of DataFrame
```

```python
# However, the above comparison is NOT true for all elements
# It is Not true in some cases. Lets first find those cases.

college_self_compare.all()
```

```python
# This is because of missing values (NaN), which are not comparable to one another
# lets see how many missing values trun up in an elementwise comparison of the whole dataFrame
# If you tried == comparisons and then summung up, you would get 0 for each variable
# This is because the primary way of counting is to use isnull method

(college_ugds_ == np.nan).sum()
```

```python
# Now lets try to use isnull dataFrame method
college_ugds_.isnull().sum()
```

```python
# So how do we proceed?
# THE CORRECT WAY OF COMPARING ENTIRE DATAFRAMES ==> use "equals" method

college_ugds_.equals(college_ugds_)

```

# Transforming the Direction of DataFrame operation

## All DataFrame methods have an * *axis* * parameter
###          The  *axis* parameter can take value *index* (rows) or *Columns*; or 0/1
###           The default value of *axis* parameter is 0 i.e. index (rows)
####           **To remember: 1 looks like a column; 
####           **read 0 as "z-row" to remind you to link zero to rows

```python
college = pd.read_csv('data/college.csv', index_col = 'INSTNM')
college_ugds_ = college.filter(like = 'UGDS_')
college_ugds_.head()
```

```python
# Counting the non-missing values in each variable -- use DataFrame operations
# The default for axis parameter is 0 or index
# So it will apply the count method across rows (for each columns) 

college_ugds_.count()

# We could have done this operations with either one of the following: 
#    college_ugds_.count(axis=0) 
#    college_ugds_.count(axis= 'index')
```

```python
# Lets trnasform the direction of DataFrame operations
# Suppose we want to check for each university, how many non-missing values exist for racial proportions  
# This will involve settin axis='columns' to transpose the operation for counting across columns

college_ugds_.count(axis='columns').head()

# We can see below that each university (row) has 9 non-missing values
```

```python
# For each university, the sum of racial proportions must add up to 1. 
# Let's check this. Need to sum across columns
# set the axis parameter to 'columns'

college_ugds_.sum(axis= 'columns').head()

```

```python
# For each university, we can vreate cumulative sum of racial proportions
# This should result in last column showing a cumulative sum of nearly 1.00

college_ugds_cumsum = college_ugds_.cumsum(axis=1)
college_ugds_cumsum.head()

# Note that the last column 'UGDS_UNKN' is nearly 1.00 for all universities
```

```python
pd.read_csv('data/college_diversity.csv', index_col='School')
```

```python
'''
We define diversity as the count of the number of races at campus having 
greater than 15% sof the student population.
We will continue to use data "college.csv"

Step 1: Check how many colleges have missing values for 1 or more race variables 
        >>chain isnull(), sum(), sort_values() methods)
        
Step 2: Drop rows with all nine race variables missing
        >>use dropna(how='all')

Step 3: Test whether a given cell has race proportion greater than 15%
        >>use .ge(.15) method

Step 4: Sum the proportions across all race variables
        >> .sum(axis='columns') method
'''

college = pd.read_csv('data\college.csv', index_col = 'INSTNM')
college_ugds_ = college.filter(like='UGDS_')
```

```python
# Step 1: Identifying the number of race variables with missing values
college_ugds_.isnull()
                      
```

```python
college_ugds_.isnull() \
                      .sum(axis=1) \
                      .sort_values(ascending=False) \
                      .head()

# As seen below, some universities have all 9 variables missing
```

```python
# Step 2: Drop rows with all race variables missing

college_ugds_ = college_ugds_.dropna(how='all')
college_ugds_.isnull().sum(axis=1).head()

# As can be seen below, the number of missing variables is zero for all universities 
```

```python
# Which cells satisfy the condition that racial proportion is greater than 15%?

college_ugds_.ge(.15)

# We create a table of booleans
```

```python
# Count the booleans for each row
diversity_metric = college_ugds_.ge(.15).sum(axis='columns')
diversity_metric.head()
```

```python
# Diverity is a single series. Confirming.
diversity_metric.shape

# Series length is 6874, as shown below. 
```

```python
# The series diversity_metric gives for each university the number of racial groups
# We want to construct the frequency distribution of this variable
# use value_counts

diversity_metric.value_counts()

# The results below show that 3042 universities have only one racial group (no diversity)
# only 2 universities have five racial groups on campus
```
